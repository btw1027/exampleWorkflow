```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def directoryFiles(dirpath):
    allfiles=os.listdir(dirpath)
    return [dirpath + '/'+ filename for filename in allfiles if filename.endswith('csv') ]
```

```python
try: 
    f=open('.dataFolderPath')
    dataFolderPath=f.read()
    f.close()
except:
    from workflowHelpers import helpSetPath
    dataFolderPath=helpSetPath.setPath()
    
dataFolderPath
```

```python
dataFiles=directoryFiles(dataFolderPath)
tempMatrix=np.zeros((1024,1024,len(dataFiles)))

for k in range(len(dataFiles)):
    reader=pd.read_csv(dataFiles[k],header=None)
    tempMatrix[:,:,k]=pd.DataFrame.as_matrix(reader)
```

```python
plt.imshow(tempMatrix.std(axis=-1))
plt.colorbar()
plt.title('Readout Noise (std. dev. of '+str(tempMatrix.shape[-1])+' dark-fast reads)')
plt.savefig('ReadNoise.png')
plt.show()

plt.imshow(tempMatrix.mean(axis=-1))
plt.colorbar()
plt.title('Baseline Counts (mean of '+str(tempMatrix.shape[-1])+' dark-fast reads)')
plt.savefig('Baseline.png')
plt.show()
```
